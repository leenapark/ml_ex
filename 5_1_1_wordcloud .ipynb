```python
from wordcloud import WordCloud
from konlpy.tag import Komoran
import matplotlib.pyplot as plt
from konlpy.corpus import kolaw   
```

```python
# 대상 텍스트 데이터 
kolaw.fileids()                          
para = kolaw.open('constitution.txt').read()
```

```python
# Komoran: 형태소 분석기
tagger = Komoran()
# 명사만 추출
nouns = tagger.nouns(para)
```

```python
nouns[:5]
```

```python
# 리스트를 문자열로 바꾸기
# text = ", ". join(nouns)
text = " ". join(nouns)
text[:8]
```

```python
# 클라우딩 객체 설정
my_path='Fonts/NanumGothic.ttf'
wordcloud = WordCloud(font_path=my_path, 
                      background_color='white', 
                      width=1200, 
                      height=800).generate(text)
fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```

### 그림을 아래와 같이 넘파이 배열로 데이터를 변환시킴

```python
import numpy as np
from PIL import Image
```

```python
icon = Image.open('images/korea_map.png').convert("RGBA")
# R, G, B : 각각의 숫자의 범위는 0~255
mask = Image.new("RGB", icon.size, (255,255,255))
mask.paste(icon, icon)
mask = np.array(mask)
```

```python
icon
```

```python
mask
```

### 그림을 배경으로 워드 클라우딩을 그기기

```python
my_path='fonts/NanumMyeongjoExtraBold.ttf'
wordcloud = WordCloud(font_path=my_path, 
                      background_color='white', 
                      width=240, 
                      height=200,
                      mask=mask).generate(text)
fig = plt.figure(figsize=(14,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```
