```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

유방암 환자 데이터셋 로드  
sklearn의 내장 dataset 활용 : load_breast_cancer  
target:
- 1 양성종양 : 착한 종양(Positive)
- 0 악성종양 : 나쁜 종양(Negative)


```python
cancer = load_breast_cancer()
```

```python
cancer
# data 값의 대응하는 target : y값
```

```python
cancer.keys()
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
print(cancer["DESCR"])
```

```python
# feature(X), target(y), feature_names 분리
data = cancer["data"]
target = cancer["target"]
feature_names = cancer["feature_names"]
```

```python
print(data)
print("-"*60)
print(target)
print("-"*60)
print(feature_names)
```

##### 데이터 프레임으로 생성

```python
df = pd.DataFrame(data=data, columns=feature_names)
df
```

```python
df["target"] = cancer["target"]
df
```

```python
df_pos = df.loc[df["target"] == 1]
df_neg = df.loc[df["target"] == 0]
```

```python
df_pos
```

```python
df_neg
```

```python
print(df_pos.shape, df_neg.shape)
```

## input 데이터 준비

```python
X = df.drop("target", axis=1) # features
y = df["target"]
```

## train, test 데이터 분할

```python
# test_size = 0.25,  shuffle=True(default)
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=20, stratify=y, test_size=0.25, shuffle=True)
```

## 모델 생성 및 학습

```python
model_dtc = DecisionTreeClassifier(max_depth=5)
model_dtc.fit(train_X, train_y)
```

## 학습 모델 테스트

```python
y_pred = model_dtc.predict(test_X)
y_pred
```

## 성능 평가

```python
print("모델의 정확도 : %.2f" % (y_pred == test_y).mean())
```

### 정확도 확인 함수 sklearn.metrics.accuracy_score(y_true, y_pred)
- sklearn.metrics.accuracy_score(y_true, y_pred)
- https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html?highlight=accuracy_score#sklearn.metrics.accuracy_score

```python
from sklearn.metrics import accuracy_score
print("모델의 정확도 : %.2f" % accuracy_score(y_pred, test_y))
```

```python
# 테스트 데이터(X_test)와 labels(y_test)로 모델의 정확도를 계산
print("모델의 정확도 : %.2f" % model_dtc.score(test_X, test_y))
```

## 성능평가 함수 설명
- from sklearn.metrics import classification_report
- https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html?highlight=classification_report

```python
from sklearn.metrics import classification_report
```

```python
# Predict를 수행하고 classification_report() 결과 출력
y_pred = model_dtc.predict(test_X)
print(classification_report(y_pred, test_y, target_names=cancer.target_names))
```

```python
print(classification_report(y_pred, test_y, labels=[0, 1]))
```

```python
print(classification_report(y_pred, test_y))
```

#### macro avg, weighted avg라는 값이 출력이 된다. 각각 아래와같은 의미를 가진다

- macro avg:
    - 각 클래스에 대한 f1-score(precision, recall)를 계산한 뒤 단순한 산술 평균을 내어 값을 계산함.
- weighted avg:
    - 불균형한 데이터일 경우에 보는 수치값.
    - 각 class에 해당하는 data의 개수에 가중치를 주어 평균을 구한 것

## 정확도의 모순
- 정확도(Accuracy) : 실제 데이터에서 예측 데이터가 얼마나 같은지 판단하는 자료
- 정확도(Accuracy) = (예측 결과와 실제 데이터가 동일한 건수) / (전체 예측 데이터 건수)

정확도(accuracy)만 보고 분류기의 성능을 판별하는 것은 **위와 같은 오류**에 빠질 수 있음.
